```python
#import requierd libraries 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
#import needed libraries for k-medoid clustering
from sklearn_extra.cluster import KMedoids
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score #Compute the mean Silhouette Coefficient of all samples
```

```python
#import needed libraries for hierarchical clustering
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as sch
```

```python
#we are going to read the data from excel file 
data = pd.read_csv('OnlineRetail.csv', sep=",", encoding="ISO-8859-1", header=0)
data
```

```python
#first we need to do preprocessed data to clean it and to make sure that the dataset is ready for clustring
```

```python
#this method let us know if there is any null values or not 
data.isnull().sum()
```

```python
#we need to remove all rows that have null values in CustomerID
my_data = data.dropna(subset=['CustomerID'])
my_data
```

```python
#to make sure that there isn't any null values 
my_data.isnull().sum()
```

```python
my_data.describe()
```

```python
#we need to delete un important columns that we don't need
my_data.drop('Country', axis = 1, inplace = True)
my_data.drop('Description', axis = 1, inplace = True)
my_data.drop('StockCode', axis = 1, inplace = True)
my_data
```

```python
#we will convert InvoiceDate to datetime type
my_data['InvoiceDate'] = pd.to_datetime(my_data['InvoiceDate'],format='%d-%m-%Y %H:%M')#day-month-year hour-min
```

```python
#we want to know the last date of transaction so we will search for maximum date
max_date = max(my_data['InvoiceDate'])
```

```python
#use this max_date to calculate difference
my_data['Difference'] = max_date - my_data['InvoiceDate']
```

```python
#RFM (recency, frequency, monetary value)
#this is the first step of RFM 
recency = my_data['Difference'].groupby(my_data['CustomerID']).min() #to group 2 attributes
recency = recency.reset_index()
recency
```

```python
#determine days only
recency['Difference'] = recency['Difference'].dt.days
recency
```

```python
#this is the second step of RMF
my_data['Amount'] = my_data['Quantity']*my_data['UnitPrice'] #any amount that we want to know = num of elements*price
monetary = my_data.groupby('CustomerID')['Amount'].sum() #to group 2 attributes 
monetary = monetary.reset_index()
monetary
```

```python
#third step of RMF
frequency = my_data['InvoiceNo'].groupby(my_data['CustomerID']).count() #count: method help us to count num of transactions
frequency = frequency.reset_index()
frequency.columns = ['CustomerID', 'Frequency']
frequency
```

```python
#here we want to merge monetary with frequency so it will be easy for us to merge the 3  attributes together
merging = pd.merge(monetary, frequency, on='CustomerID', how='inner')
merging
```

```python
#so the RMF is to merge recency, frequency, monetary together 
RMF = pd.merge(merging, recency, on = 'CustomerID', how = 'inner')
RMF.columns = ['CustomerID', 'Amount', 'Frequency', 'Recency']
RMF
```

```python
sns.boxplot(RMF['Amount'])
```

```python
#this is to remove outliers 
#IQR
Q1 = np.percentile(RMF['Amount'], 25, interpolation = 'midpoint')  #25% for Q1
Q3 = np.percentile(RMF['Amount'], 75, interpolation = 'midpoint')  #75% for Q3
IQR = Q3 - Q1
#Upper bound
upper = np.where(RMF['Amount'] >= (Q3+1.5*IQR)) #condition to remove outliers
#Lower bound
lower = np.where(RMF['Amount'] <= (Q1-1.5*IQR))
RMF.drop(upper[0], inplace = True)
RMF.drop(lower[0], inplace = True)
```

```python
sns.boxplot(RMF['Amount'])
```

```python
sns.boxplot(RMF['Frequency'])
```

```python
sns.boxplot(RMF['Recency'])
```

```python
RMF
```

```python
scaler = StandardScaler().fit(RMF)  #mean removal and variance scaling
x_scaled = scaler.transform(RMF) #Perform standardization by centering and scaling
```

```python
sw = []
#Silhouette Width is the estimate of the average distance between clusters
#sw values are between 1 and -1 
#the value 1 indicates a good cluster
for i in range(2, 9):
    kMedoids = KMedoids(n_clusters = i, random_state = 0)
    kMedoids.fit(x_scaled)
    y_kmed = kMedoids.fit_predict(x_scaled)
    silhouette_avg = silhouette_score(x_scaled, y_kmed)
    sw.append(silhouette_avg)
```

```python
plt.plot(range(2, 9), sw)
plt.title('Silhouette Score')
plt.xlabel('Number of clusters')
plt.ylabel('SW')      #within cluster sum of squares
plt.show()
```

```python
#the sw should be high so the distance between clusters will be high 
#so we will choose num of clusters = 4
```

```python
cluster = KMedoids(n_clusters=4, metric='manhattan',random_state = 0) 
cluster.fit_predict(RMF) #fitting the data and return lable as it is a unsupervised learning
```

```python
#if we want to know which centers we choose:
cluster.cluster_centers_
```

```python
data = RMF.iloc[:, 1:4].values
plt.scatter(data[:,0], data[:,1], data[:,2], c=cluster.labels_, cmap='rainbow')
```

```python
RMF['label_kmedoid'] = cluster.fit_predict(RMF) #to make a new column which will be target column or clusters 
```

```python
#ploting a dendrogram to determine optimal number of clusters
dendrogram = sch.dendrogram(sch.linkage(RMF,  "single"))
```

```python
# the clustering step and creating model
Cluster = AgglomerativeClustering(n_clusters=2,affinity='euclidean', linkage='single')
Cluster.fit_predict(RMF)
```

```python
#ploting the data with each cluster
plt.scatter(data[:,0], data[:,1], data[:,2], c=Cluster.labels_, cmap='rainbow')
```

```python
#making a new column which containes the cluster for each instance
RMF['label_hierarchical'] = Cluster.fit_predict(RMF) 
```
